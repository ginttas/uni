````python
import numpy
import math
import matplotlib.pyplot as plt
````

Define x array

````python
start = 5
end = 20
n = 2000
x = numpy.linspace(start, end, n)
#display(x)
````

Calculate y array

````python
def func(x):
    y = x * abs(math.cos(x))
    return y
y = [func(i) for i in x]
#print(y)
````

Plot x vs y

````python
plt.plot(x, y)
````
