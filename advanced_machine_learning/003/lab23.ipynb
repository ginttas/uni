```python
import pandas

df = pandas.read_csv("../data/owid-covid-data.csv")
df
```

```python
df.describe()
```

```python
dff = df[["date", "excess_mortality", "human_development_index"]]
dff = dff.dropna()
date = dff["date"].value_counts().index[0]
dff = dff[dff["date"] == date]
dff
```

```python
import matplotlib.pyplot as plt

dff = dff[dff["excess_mortality"] < 50]

plt.plot(dff["human_development_index"], dff["excess_mortality"], "o")
```
