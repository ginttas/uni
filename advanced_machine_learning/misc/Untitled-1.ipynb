#### Jonas Matuzas: Dirbtinio intelekto taikymai laužynuose (Ovoko)

Ovoko - vienas izdziausiu naudotu autombiliu deltaliu pardavieju lietuvoje (pleciasi i EU)
Svarbi tavurumo ideja

ML skirtas surasti kokias detales galime utilizuoti (kas nebus pardodoams, o ka galima)

Lauzyno darbotojes ivesdamas detale turi suvesti apie ja visa imformacija (kas tai, kam naudojema, kokia masina)

Computer vision naudojemas - Automatiskai atlikti sios vei,msmus (aka. sukatalizuoti)

Sunkus uzdavinys

Ant detaliu gali buti nusedevieje numeriai (aprudyjusios, skirtingi uzrasymasi (ispausti, uzrasyti, liptukai ir t.t., kaip buvo fotiknita ir t.t.)))

Dar neaisku kurios numerius pasirinkti nes jus gali buti belkiek ant vienos detales

Su 6 modeliu paglba pavyko tai padaryti

1. Pirmas modelis skirtas - Angle detectiom Image rotation (Nutraukos gera kryptis)
2. Iekoma numeriu (Bound Box)
3. OCR Nuskaityti numneri
4. Nustatyti kuris skaiciukas yra geras (Pagal numerio struktura) (Numeris keliauje i db, bet jus vis atsiranda tai visuis turieti viename db labai sunku)
5. Jeigu nera numerio, tai kategorija nustatoma is vaizdo
6. Is paveiksliuko nustatyti gamintoja

##### Is paveikliuko gamintojo atpazinimas
* Jau buvo daug istoriniu duomenu - paveiksliuku su lables
* Buvo nubraizuyta konfusion matrica
* Kai kurios kompnijos turi tas patces detales (is konfusion matricos matom,o kurios kompanijos suisijungusios) (kas nauydoje panases detales)

##### Kategoriju detekcija
* Isukis buvo nes 1600 kategorijos
* Bet padejo apsiribojent kategorijom kurios dzaniausei neturi partnumberiu 350
