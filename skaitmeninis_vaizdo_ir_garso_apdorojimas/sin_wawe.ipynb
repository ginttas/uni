```python
%matplotlib widget
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(w, amp, phi):
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.set_ylim([-4, 4])
    ax.grid(True)
    
    # generate x values
    x = np.linspace(0, 3 * np.pi, 1000)
    
    def my_sine(x, w, amp, phi):
        """
        Return a sine for x with angular frequeny w and amplitude amp.
        """
        return amp*np.sin(w * (x-phi))
    
    ax.plot(x, my_sine(x, 1.0, 1, 0), color='C1')
    ax.plot(x, my_sine(x, w, amp, phi), color='C0')
    plt.show()
```

```python
plot(w = 2, amp = 4, phi = 0)
```

```python
#w, amp, phi
def my_sine(x, w, amp, phi):
    """
    Return a sine for x with angular frequeny w and amplitude amp.
    """
    return amp*np.sin(w * (x-phi))

def tmp(x, par):
    y = sum([my_sine(x, i[0], i[1], i[2]) for i in par])
    return y

def complex(waves):
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.set_ylim([-4, 4])
    ax.grid(True)
    x = np.linspace(0, 3 * np.pi, 1000)
    y = [tmp(i, waves) for i in x]
    ax.plot(x, y, color='C0')

    dy = [((y[i+1] - y[i])/(x[i+1] - x[i])) for i in range(len(x) - 1)]
    dx = x[:-1]
    ax.plot(dx, dy, color='C1')

    ddy = [((dy[i+1] - dy[i])/(dx[i+1] - dx[i])) for i in range(len(dx) - 1)]
    ddx = dx[:-1]
    ax.plot(ddx, ddy, color='C2')

    dddy = [((ddy[i+1] - ddy[i])/(ddx[i+1] - ddx[i])) for i in range(len(ddx) - 1)]
    dddx = ddx[:-1]
    #ax.plot(dddx, dddy, color='C3')

    plt.show()

    
```

```python
# w, amp, phi
complex([
    [7, 0.1, 0.5],
    [4.5, 0.05, 0],
    [-2, 0.05, 0.5],
    [5, 0.05, 0.5],
])
```
