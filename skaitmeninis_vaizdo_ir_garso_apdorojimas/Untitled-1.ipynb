### The nature of Sound and Visual images

#### The nature of sound
Grasas - akutisnes/mechanies bangos kurios sklida kazkokia terpe (dujos, skystis, kieta) - Perdodma enrgija dalelemis - Garsas gasli atssipindieti - Luzti - Buti apsorbotas priklausomai nuo terpes

Pvz.: Diesttyojes sneka - Mikrofonas paguanba - Garas i signala .. - VAP faials (amplitudziu reiksmes kaip mikrofono membrans buvo pasislinkus, koks buvo sliegis)

Is dumenu strukturos garsas - vienmatis vektorius (Konkreciu laiko moduliu buvo mikrafono deformacija)

#### Fizines charakteristimos

Garsas kaip banga - matematinsi apdorojimas (sin, cos, ampiltude, daznis, pitch (storas/zemas))

Violencele groje zemaiu, negu smuikas, kotrabosas dar zemiau

Amplitude -garo stiprumas

Faze - Dar grysim

Judejimo greicio terpeja (dar grysim)

Attenulatioin (Prislopimas) - Kaip energija isnyksta dauzydamis i molekules

#### Bangos ilgis ir dazinsi
Skirti aprasyti perijodiuskai kartojentems procesams aprasyti

* E(t) = A sin(2 pi f t + fi)
* A - amplitude
* f - dazinis
* t - laikas
* fi - faze

* f = v/A
* A - bangos igis
* v - greitis

* f = 1/T
* T - perijodas

* A - Tv

#### Amplitude
Kai "Deformatas oras pasieke musu ausis" musu ausu mebrana suda,
didelis sliegis - juda i vidu
mazas - i apacia

Amplitude yra logaritmine skale (matojema dB)

Zmogaus ausies skalew yra logaritmine

#### Daznis
GIRDIMAS garasas 20hz to 20khz
Irig naudojema logaritmine sakle

Gitaros styga ar kito instrumento - pradeda vibroti (Formoje akustine banga) styga vibroje visu savo ilgiu, pusia ilgio ir t.t.

Styga genroje:
Bazini dazni,
Harmoninkas
ir papildomu overtonu del instumentu savybiu

(Skirtingi garasai tarp instumentu atsirada del Overtonu/tempru/natos spalvos)
